```python
import pickle
data = pickle.load(open("recipe_ingredient_bows.pkl", "rb"))
```

```python

all_ing = set()
for row in data:
    all_ing.update(row)
print(len(all_ing))
```

```python
ing_list = [a for a in all_ing]
for i in range(len(ing_list)):
    for j in range(len(ing_list)):
        if not i == j:
            if len(ing_list[i]) > len(ing_list[j]):
                if ing_list[i].endswith(" " + ing_list[j]):
                    print(ing_list[i] + "," + ing_list[j])
            elif len(ing_list[i]) < len(ing_list[j]):
                if ing_list[j].endswith(" " + ing_list[i]):
                    print(ing_list[j] + "," + ing_list[i])
```
